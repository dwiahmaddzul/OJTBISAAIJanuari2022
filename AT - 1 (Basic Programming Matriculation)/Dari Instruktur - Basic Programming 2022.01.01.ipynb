**While loop**

```python
i = 1

while i < 11: #selama i kurang dari 10
    print(i) # print i
    i = i + 1 # tambahkan +1 setiap iterasi
```

```python
n = 1

while n < 5:
    print(n)
```

```python
# while loop - if
# angka ganjil/genap : habis di bagi 2
# gunakan operator 'remainder', yaitu '%'
# 10 % 5 == 0 (sisa bagi)

x = 1

while x < 11 :
    if x % 2 == 0:
        print( str(x) + ' bilangan genap')
    else:
        print( str(x) + ' bilangan ganjil')
    x += 1
```

**for loop**

```python
words = ['hello','world','spam','eggs']
```

```python
for i in words: # i = iterasi index dalam list word
    print(i)
```

```python
for i in range (1,11):
    print(i)
```

```python
# for loop - if

for i in range (1,11):
    if i % 2 == 0:
        print( str(i) + ' bilangan genap')
    else:
        print( str(i) + ' bilangan ganjil')
```

## function
def & class

```python
def penjumlahan(a,b):
    hasil = a + b
    return hasil
```

```python
penjumlahan(4,5)
```

```python
def perkalian (c,d):
    hasil = c*d
    return hasil
```

```python
perkalian (2,8)
```

```python
# class

class operasi ():
    
    def penjumlahan(a,b):
        hasil = a + b
        return hasil

    def perkalian (c,d):
        hasil = c*d
        return hasil
```

```python
operasi.penjumlahan(2,4)
```

```python
operasi.perkalian ( 6,7)
```
