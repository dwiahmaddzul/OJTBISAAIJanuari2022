### Model Visualisasi

#### Persiapan Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from wordcloud import WordCloud,STOPWORDS
from pywaffle import Waffle
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

#### Visualisasi : Piechart dan Countplot

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Iris Species Count')
ax[0].set_ylabel('Count')
sns.countplot('Species',data=df,ax=ax[1])
ax[1].set_title('Iris Species Count')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
df["PetalWidthCm"].groupby(df["Species"]).mean().plot.pie()
plt.subplot(2,2,2)
df["SepalWidthCm"].groupby(df["Species"]).mean().plot.pie()
plt.subplot(2,2,3)
df["SepalLengthCm"].groupby(df["Species"]).mean().plot.pie()
plt.subplot(2,2,4)
df["PetalLengthCm"].groupby(df["Species"]).mean().plot.pie()
```

#### Visualisasi : Histogram

```python
df.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,6)
```

#### Visualisasi : Heatmap

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.heatmap(df.corr(),annot=True,cmap='coolwarm',linewidths=1,linecolor='k',square=True,mask=False, vmin=-1, vmax=1,cbar_kws={"orientation": "vertical"},cbar=True)
```

#### Visualisasi : Wordcloud

```python
datawordcloud = pd.read_csv("Scrapping_Bibit.csv")

plt.figure(figsize = (20,20)) 
wc = WordCloud(max_words = 100 , width = 1600 , height = 800).generate(" ".join(datawordcloud.content))
plt.imshow(wc , interpolation = 'bilinear')
```

#### Visualisasi : Waffle

```python
dfwaffle = df
dfwaffle = dfwaffle.groupby('Species').size().reset_index(name='counts')
dfwaffle
```

```python
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5,
    values=dfwaffle.counts,
    labels=list(dfwaffle.Species),
    figsize=(15, 15)
)

plt.title('Distribusi Spesies')

plt.show()
```
