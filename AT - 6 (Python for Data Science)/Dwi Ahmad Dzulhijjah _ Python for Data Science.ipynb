##### Piechart

```python
df['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
df["PetalWidthCm"].groupby(df["Species"]).mean().plot.pie()
plt.subplot(2,2,2)
df["SepalWidthCm"].groupby(df["Species"]).mean().plot.pie()
plt.subplot(2,2,3)
df["SepalLengthCm"].groupby(df["Species"]).mean().plot.pie()
plt.subplot(2,2,4)
df["PetalLengthCm"].groupby(df["Species"]).mean().plot.pie()
```
