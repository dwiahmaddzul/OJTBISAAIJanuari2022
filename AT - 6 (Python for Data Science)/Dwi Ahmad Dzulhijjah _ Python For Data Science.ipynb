```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight') 
import warnings
warnings.filterwarnings('ignore') 

import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
data.head(10)
```

#### SOAL : VISUALISASI DATA

#####  Visualisasi Bar Plot dan Pie Chart

```python
data.describe().plot(kind = "area", figsize = (15,8), colormap="turbo")
plt.xlabel('Statistics',fontsize=14)
plt.ylabel('Value',fontsize=14)
plt.title("General Statistics of Iris Dataset",fontsize=18)
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Iris Species Count')
ax[0].set_ylabel('Count')
sns.countplot('Species',data,ax=ax[1])
ax[1].set_title('Iris Species Count')
```

```python
ax=plt.subplots(1,2,figsize=(18,8))
data['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
plt.show()
```
