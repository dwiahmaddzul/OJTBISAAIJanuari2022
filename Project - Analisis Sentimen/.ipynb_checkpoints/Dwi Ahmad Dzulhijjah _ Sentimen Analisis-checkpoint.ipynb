### Persiapan Data

#### Preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv", sep = ',')

df['sex'] =df['sex'].astype('category')
df['region'] =df['region'].astype('category')
df['smoker'] =df['smoker'].astype('category')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(df.sex.drop_duplicates())
df.sex = label.transform(df.sex)
label.fit(df.smoker.drop_duplicates())
df.smoker = label.transform(df.smoker)
label.fit(df.region.drop_duplicates())
df.region = label.transform(df.region)
df.dtypes
```

```python
def Bmi_Classifier(row):
    if row['bmi'] < 18.5:
        return "underweight"
    elif row['bmi'] < 25:
        return "normal"
    elif row['bmi'] < 30:
        return "overweight"
    else:
        return "obese"

def Age_Classifier(row):
    if row['age'] < 17:
        return "child"
    elif row['age'] < 31:
        return "adult"
    elif row['age'] < 46:
        return "middle-aged"
    else:
        return "old-aged"

df['bmi_status'] = df.apply(Bmi_Classifier, axis=1)
df['age_status'] = df.apply(Age_Classifier, axis=1)

df.head()
```

### Machine Learning Modelling

#### Linear Regression

```python
data2 = df[df['smoker'] == 'yes'][['sex', 'children', 'region', 'charges','bmi_status', 'age_status']]
new_data = pd.get_dummies(data2, columns = ['sex','children','region','bmi_status','age_status'])
new_data.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = new_data.drop('charges', axis = 1).values
y = new_data['charges']

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.80, test_size = 0.20, random_state = 15)

lm = LinearRegression(fit_intercept = True)
lm.fit(X_train, y_train);

y_pred = lm.predict(X_train)
```

#### MATRIKS EVALUASI

```python
from sklearn.metrics import r2_score, explained_variance_score, mean_absolute_error, mean_squared_error
from math import sqrt
```

```python
print('Akurasi data latih: ',lm.score(X_train, y_train).round(3))
print('Akurasi data Uji: ',lm.score(X_test, y_test).round(3))
print(' ')
print('Root Mean Squared Error (RMSE)')
print('RMSE data latih: ', round(sqrt(mean_squared_error(y_train, y_pred)),3))
print('RMSE data uji: ', round(sqrt(mean_squared_error(y_test, lm.predict(X_test))),3))
print(' ')
print('Mean Absolute Error (MAE)')
print('MAE data latih: ', mean_absolute_error(y_train, y_pred).round(3))
print('MAE data uji: ', mean_absolute_error(y_test, lm.predict(X_test)).round(3))
```

### PREDIKSI

#### Uji Coba Prediksi

```python
X2 = new_data.drop('charges', axis = 1)

Inter = lm.intercept_
Coeff = lm.coef_
Coeff = Coeff.reshape(-1,19)

Coeff_df = pd.DataFrame(Coeff, columns = [X2.columns])
Coeff_df
```

```python
sex = ['sex_female']
children = ['children_2']
region = ['region_southwest']
bmi_status = ['bmi_status_normal']
age_status = ['age_status_middle-aged']

pred1 = (Inter) + (Coeff_df[sex].values[0] * 1) + (Coeff_df[children].values[0] * 1) + (Coeff_df[region].values[0] * 1) + (Coeff_df[bmi_status].values[0] * 1) +(Coeff_df[age_status].values[0] * 1)

print('Prediksi Biaya Asuransi Anda Adalah : ', pred1[0].round(2), 'dollars')
```
